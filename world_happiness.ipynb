```python
import pandas as pd 

df = pd.read_csv('WHRFinal.csv')
```

```python
# creating 
df.columns = df.columns.str.strip()
```

```python
df.describe()
```

```python
# to fill the missing_value

mean_value = df['Life Expectancy'].mean()

df['Life Expectancy'] = df['Life Expectancy'].fillna(mean_value)
```

```python
df.columns = df.columns.str.lower()

df.columns = df.columns.str.replace(' ','_')
```

```python
df.info()
```

```python
# creatting new column (Happiness_Level)

labels = ['Low', 'Medium', 'High']

df['happiness_level'] = pd.qcut(df['happiness_score'], q=3 , labels= labels)

```

```python
df[['happiness_rank', 'happiness_level']].head(20)
```

```python
# creatting new column (INCOME_GROUP)
label = ['low_income', 'middl_income', 'high_income']

df['income_group'] = pd.qcut(df['gdp'], q=3 , labels=label)
```

```python
df.to_csv("cleane_happiness.csv", index=False)
```
